```python
# Initialize an empty list to hold each student's dictionary containing their name and average score
student_score_list = []

# Prompt the user for the number of students
while True:
    try:
        num_of_students = int(input('Enter the Number of Student: ').strip())
        break  # Exit loop if valid input
    except ValueError:
        print('Kindly Enter an Integer')  # Notify user on invalid input

# Prompt the user for the number of subjects
while True:
    try:
        num_of_subject = int(input('Enter the Number of Subject: ').strip())
        break  # Exit loop if valid input
    except ValueError:
        print('Kindly Enter an Integer')  # Notify user on invalid input

# Loop through each student based on the total number of students
for stu in range(1, (num_of_students + 1)):
    individual_student_score_list = []  # List to store the scores of this student

    # Prompt for and validate the student's name
    while True:
        try:
            stu_name = input(f'Enter student {stu} name : ')
            if not stu_name.isalpha():  # Name should only contain letters
                raise ValueError
            break
        except ValueError:
            print('Names should be only letters')

    # Loop through each subject for the current student
    for stu_sub in range(1, (num_of_subject + 1)):
        while True:
            try:
                name_of_sub = input('Enter name of subject: ')  # Subject name is input (but not used)
                subject_score = int(input(f'Enter {stu_name.title()} scores in {name_of_sub.title()}: '))  # Input score

                if subject_score > 100:
                    raise ValueError  # Raise error if score is above 100

                individual_student_score_list.append(subject_score)  # Add valid score to the list
                break
            except ValueError:
                # The variable subject_score might be undefined if the error is due to conversion
                print('Score is invalid. It must be an integer and not more than 100.')


    # Add this student's dictionary to the main list
    student_score_list.append({
        'Name' : stu_name,
        'Scores' : individual_student_score_list,
        'Avg_scores' : sum(individual_student_score_list) / len(individual_student_score_list)
    })

# Loop through each student dictionary to assign grades based on their average score
for student_each in student_score_list:
    if 80 <= student_each['Avg_scores'] <= 100:
        print(f'\nHi {student_each["Name"].title()} you scored an average of {round(student_each["Avg_scores"],2)} your grade is A, excellent')
    elif 70 <= student_each['Avg_scores'] <= 79:
        print(f'\nHi {student_each["Name"].title()} you scored an average of {round(student_each["Avg_scores"],2)} your grade is B, very good')
    elif 60 <= student_each['Avg_scores'] <= 69:
        print(f'\nHi {student_each["Name"].title()} you scored an average of {round(student_each["Avg_scores"],2)} your grade is C, Good')
    elif 50 <= student_each['Avg_scores'] <= 59:
        print(f'\nHi {student_each["Name"].title()} you scored an average of {round(student_each["Avg_scores"],2)} your grade is D, Fair Needs improvement')
    elif student_each['Avg_scores'] < 50:
        print(f'\nHi {student_each["Name"].title()} you scored an average of {round(student_each["Avg_scores"],2)} your grade is F, Work Harder')
```
